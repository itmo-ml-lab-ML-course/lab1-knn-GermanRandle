```python
%run "02-algorithm.ipynb"

from IPython.display import clear_output


def my_accuracy(k, X, y):
    predictions = KNN(k, WindowType.VARIABLE, Kernel.TRIANGULAR, Metric.MANHATTAN).fit(X_train, y_train, default_weights).predict(X)
    return accuracy(y.to_list(), predictions)


def lib_accuracy(k, X, y):
    predictions = KNeighborsClassifier(n_neighbors=k, weights='uniform', algorithm='ball_tree', p=2.5).fit(X_train, y_train).predict(X)
    return accuracy(y.to_list(), predictions)


clear_output(wait=False)
```

```python
xs = [x for x in range(1, 201)]
ys_my_val, ys_my_test, ys_lib_val, ys_lib_test = [], [], [], []
for k in xs:
    ys_my_val.append(my_accuracy(k, X_val, y_val))
    ys_my_test.append(my_accuracy(k, X_test, y_test))
    ys_lib_val.append(lib_accuracy(k, X_val, y_val))
    ys_lib_test.append(lib_accuracy(k, X_test, y_test))
    print(k, "done")

plt.plot(xs, ys_my_val, label='my, val')
plt.plot(xs, ys_my_test, label='my, test')
plt.plot(xs, ys_lib_val, label='lib, val')
plt.plot(xs, ys_lib_test, label='lib, test')
plt.xlabel('k')
plt.ylabel('Accuracy')
```
